```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

vidcap = cv2.VideoCapture('aic19-track3-train-data/2.mp4')
success,image = vidcap.read()
count = 0

h,w,_ = image.shape
window_size = 5



temporal_difference = []
# get current frame
old_image = image

while success:
    success,image = vidcap.read()
    count += 1
    if count%1000 != 0:
        continue
#     print count
    
    current_difference = np.ones((h-window_size, w-window_size))
    
    for x in range(h-window_size):
        for y in range(w-window_size):
#             print x,y
            sub_img = image[x:x+window_size, y:y+window_size,:]
            current_difference[x,y] = sum(sum(sum(np.abs(sub_img - old_image[x:x+window_size, y:y+window_size,:]))))

    
    old_image = image
    temporal_difference.append(current_difference)

```

```python

temporal_difference = np.array(temporal_difference)    
temporal_difference_origin = temporal_difference.copy()

plt.imshow(temporal_difference_origin[2,:,:])
temporal_difference_origin[2,:,:]
```

```python
temporal_difference = np.array(temporal_difference)    
# Process based on temporal
static_temporal_max = 20

results = np.zeros((temporal_difference.shape[1], temporal_difference.shape[2]))

for i in range(temporal_difference.shape[1]):
    for j in range(temporal_difference.shape[2]):
        max_zeros = 0
        count = 0
        for k in range(temporal_difference.shape[0]-1):
            if(temporal_difference[k,i,j] == temporal_difference[k+1,i,j] and temporal_difference[k,i,j] == 0):
                count+=1
                continue
            max_zeros = max(max_zeros, count)
            count = 0
        max_zeros = max(max_zeros, count)
        results[i,j] = max_zeros
                                                                                                  
                                                                                                  
        
```
